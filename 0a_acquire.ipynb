If something gets stuck, run `from acquire import cleanup; cleanup("COM3")` in a new cell.

```python
from pathlib import Path
from datetime import datetime
from acquire import build_campy_config, acquire, encode_raw_videos, verify_videos
import json

date = datetime.now().strftime("%Y%m%d")
mouse_1 = "slmc001"
mouse_2 = "slmc002"
assay = "open_field"
experimenter = "IT"
cohort = "cohort_1"
cage = "cage_1"
notes = ""

base_data_dir = Path(r"C:\Users\isaac\Desktop\3dpose\data")
base_config_path = Path(r"C:\Users\isaac\Desktop\3dpose\data\configs\3dpose_6cams.yaml")
serial_port = "COM3"
trigger_pins = [2, 4, 6, 8, 10, 12]
frame_rate = 100
camera_names = ["cam1", "cam2", "cam3", "cam4", "cam5", "cam6"]

session_id = f"{mouse_1}_{mouse_2}"
session_dir = base_data_dir / date / session_id
calibration_dir = session_dir / "calibration"
recording_dir = session_dir / "recording"

for d in [calibration_dir, recording_dir]:
    d.mkdir(parents=True, exist_ok=True)

print(f"{date} | {session_id} | {assay} | {session_dir}")
```

## Calibration (optional — skip if cameras haven't moved)

```python
cfg = build_campy_config(base_config_path, calibration_dir, date, session_id, camera_names, "calibration")
acquire(cfg, serial_port, trigger_pins, frame_rate, label="calibration")
encode_raw_videos(calibration_dir, camera_names, cfg)
verify_videos(calibration_dir, camera_names)
```

## Recording

```python
cfg = build_campy_config(base_config_path, recording_dir, date, session_id, camera_names, "recording")
acquire(cfg, serial_port, trigger_pins, frame_rate, label="recording")
encode_raw_videos(recording_dir, camera_names, cfg)
verify_videos(recording_dir, camera_names)
```

## Save Metadata

```python
metadata = dict(
    date=date, session_id=session_id, mouse_1=mouse_1, mouse_2=mouse_2,
    assay=assay, cohort=cohort, cage=cage, experimenter=experimenter,
    notes=notes, num_cameras=len(camera_names), camera_names=camera_names,
    frame_rate=frame_rate, resolution=[1920, 1200],
)
meta_path = session_dir / "session_metadata.json"
with open(meta_path, "w") as f:
    json.dump(metadata, f, indent=2)
print(f"Saved: {meta_path}")
```
